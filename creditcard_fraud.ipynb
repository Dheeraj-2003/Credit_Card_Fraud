Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =  train_test_split(df.drop(columns=['Class']),df['Class'],test_size=0.2,random_state=20)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
x_train.head()
```

Scaling

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
ss = StandardScaler()
```

```python
amount_train = ss.fit_transform(x_train['Amount'].to_numpy().reshape(-1,1))
x_train['Amount'] = amount_train
x_test['Amount'] = ss.fit_transform(x_test['Amount'].to_numpy().reshape(-1,1))
```

```python
x_test
```

```python
x_train['Time'] = (x_train['Time'] - x_train['Time'].min())/(x_train['Time'].max()-x_train['Time'].min())
x_test['Time'] = (x_test['Time'] - x_test['Time'].min())/(x_test['Time'].max()-x_test['Time'].min())
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,f1_score,accuracy_score,roc_auc_score
```

```python
print(classification_report(y_test,y_pred_lr, target_names=['not fraud','fraud']))
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf',C=100)
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred_svc = svc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_svc))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
y_pred_gnb = gnb.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100,max_depth=10,n_jobs=-1)
rfc.fit(x_train, y_train)
y_pred_rfc = rfc.predict(x_test)
```

```python
print(classification_report(y_test,rfc.predict(x_test)))
```

```python
print(roc_auc_score(y_test,rfc.predict(x_test)))
```

```python
x_train.head()
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
print(classification_report(y_test,xgb.predict(x_test)))
```

```python
y_pred_xgb = xgb.predict(x_test)
```

```python
f1_score(y_test,xgb.predict(x_test))
```

```python
print(roc_auc_score(y_test,y_pred_lr))
print(roc_auc_score(y_test,y_pred_svc))
print(roc_auc_score(y_test,y_pred_gnb))
print(roc_auc_score(y_test,y_pred_rfc))
print(roc_auc_score(y_test,y_pred_xgb))
```
